## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
x = 6
#n=4x μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
# Τελευταίο ψηφίο είναι το 6.
n = 4*x
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n = 4*6, k = 4, p = 0.0, seed = seedno) # Βάζουμε πιθανότητα p=0 για το πλέγμα βαθμού 4
REG.graph["name"] = "REG"
fig, ax = plt.subplots(figsize = (10,8))
nx.draw_circular(REG, ax = ax, with_labels = True)
ax.set_title(f"Τοπολογία REG με βαθμό πλέγματος k = {d} και n = {n}")
plt.show()

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
x = 6
n=4*x 
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n = n, m = M, seed = seedno) # Set seed here
RGER.graph["name"] = "RGER"

fig, ax = plt.subplots(figsize = (10,8))
nx.draw_circular(RGER, ax = ax, with_labels = True)
ax.set_title(f"Τοπολογία RG (Erdos Renyi) με πλήθος κόμβων n={n} και πλήθος ακμών m = {M}")
plt.show()

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
x = 6
n = 4*x
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n = n, radius = R,
                                seed = seedno) # Set seed here
RGG.graph["name"] = "RGG"

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    z, y = pos[n]
    d = (z - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(10, 8))
nx.draw_networkx_edges(RGG, pos)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
)
plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.title(f"Τοπολογία Random Geometric Graph με ακτίνα σύνδεσης R={0.35} και πλήθος κόμβων n={24}")
plt.show()
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
"""Συνάρτηση που υπολογίζει τον Συντελεστή Ομαδοποίησης
    για κάθε κόμβο (Clustering coefficient), επιστρέφει ένα
    λεξικό με κλειδιά τους κόμβους και τιμές τον αντίστοιχο συντελεστή"""
def cc(G):
    coefs = nx.clustering(G)
    vals = list(coefs.values())
    return coefs, np.mean(vals), np.var(vals)

import seaborn as sns # For a better theme

sns.set_style("dark")
# A function to visualize the distribution of clustering coefficients
def visualize_cc(Graphs, coefs, w = 6, h = 4):
    fig, ax = plt.subplots(figsize = (w, h), nrows =1, ncols = len(Graphs))
    for idx, G in enumerate(Graphs):
        coef = coefs[idx][0]
        probs = list(coef.values())
        ax[idx].hist(probs)
        ax[idx].set_xlabel("Clustering Coefficient")
        ax[idx].set_ylabel("Number of Nodes")
        ax[idx].set_title(G.graph['name'])
    plt.show()
```

##### B1.1 Συντελεστής ομαδοποίησης και αναπαράσταση κατανομής

Σε αυτό το σημείο, χρησιμοποιώντας τη συνάρτηση `cc` υπολογίζουμε τον συντελεστή ομαδοποίησης για κάθε έναν απ' τους γράφους
που δημιουργήσαμε και με τη βοήθεια της συνάρτησης `visualize_cc` οπτικοποιούμε την αντίστοιχη κατανομή που προκύπτει.

```python
graphs = [REG, RGER, RGG, SF, SW]
```

```python
clustering_info = []
for G in graphs:
    clustering_info.append(cc(G))
```

```python
visualize_cc(graphs, clustering_info,w=20)
```

##### Β1.2 Μέσος συντελεστής ομαδοποίησης και η διασπορά του.

```python
for idx, G in enumerate(graphs):
    print(f"- "+ G.graph['name'] + f" with mean cc value {clustering_info[idx][1]} and variace {clustering_info[idx][2]}")
```

##### Β.1.3 Εύρεση κόμβων με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 

```python
for idx, G in enumerate(graphs):
    max_value = max(clustering_info[idx][0].values())
    min_value = min(clustering_info[idx][0].values())
    max_nodes = []
    min_nodes = []
    for node,val in clustering_info[idx][0].items():
        if val == max_value:
            max_nodes.append(node)
        if val == min_value:
            min_nodes.append(node)
    print("- " + G.graph['name'] + f" graph with max cc value: {max_value}, nodes: {max_nodes}, min cc value: {min_value}, nodes: {min_nodes}")
```

##### Σχολιασμός Αποτελεσμάτων

- <b>Τοπολογία REG:</b> Όπως βλέπουμε, στην περίπτωση του κανονικού γράφου (τοπολογία REG) η κατανομή του συντελεστή ομαδοποίησης είναι ντετερμινιστική εφόσον ο βαθμός κάθε κόμβου είναι ίσος με 4 και οι συνδέσεις γίνονται μόνο μεταξύ των 4 πλησιέστερων γειτόνων. Έτσι, για κάθε κόμβο $i$ ο αριθμός των τριπλετών που είναι συνδεδεμένοι με τον κόμβο $i$ είναι ίσος με 3. Τώρα, εφόσον ο βαθμός του κόμβου $i$ είναι ίσος με $k_i = 4$, το σύνολο όλων των τριπλετών με κέντρο τον κόμβο $i$ θα είναι ίσο με $\binom{4}{2} = 6$. Έτσι, ο συντελεστής ομαδοποίησης του κόμβου $i$ θα είναι ίσος με 

$$C_i = \frac{\text{πλήθος τριπλετών συνδεδεμένες με τον κόμβο }i}{\text{πλήθος συνολικών τριπλετών με κορυφή τον κόμβο } i} = \frac{3}{\binom{4}{2}} = \frac{3}{6} = 0.5.$$

Έτσι, για κάθε κόμβο $i$ έχουμε ότι $C_i=0.5$ επομένως η κατανομή είναι ντετερμινιστική εφόσον $\mathbb{P}(C_i = 0.5)=1$ για κάθε $i\in \{1,\dots n\}$. Πράγμα το οποίο επιβεβαιώνεται και απ' το πρώτο σχήμα κατά σειρά στην παραπάνω εικόνα. Τέλος, όπως είναι αναμενόμενο η μέση τιμή του συντελεστή ομαδοποίησης είναι επίσης $0.5$ και η διασπορά $0$.

- <b>Τοπολογία RG ($\text{Erd\H{o}s - R\'enyi}$): </b> Η τοπολογία του τυχαίου γράφου κατά $\text{Erd\H{o}s - R\'enyi}$ ισοδυναμεί με την τυχαία επιλογή ενός γράφου (subgraph) από το σύνολο $\mathcal{G}(n,m)$, όπου το σύνολο $\mathcal{G}(n,m)$ αποτελείται από τους $\binom{N}{m}$ υπογράφους με $m$ ακμές, του πλήρη γράφου $K_n$ με $n$ κόμβους και $N = \frac{n(n-1)}{2}$ ακμές. Στην περιπτώσή μας έχουμε $n=24$ και $M=150$. Με αυτά τα δεδομένα είναι εύκολο να υπολογίσουμε την πιθανότητα $p$ δύο κόμβοι να συνδέονται με κάποια ακμή στον τυχαίο γράφο που επιλέγεται. Πράγματι, το πλήθος όλων των δυνατών ακμών είναι ίσο με $\frac{24\cdot 23}{2} = 12\cdot 23 = 276$. Έτσι, η αν επιλέξουμε δύο τυχαίους κόμβους του τυχαίου γράφου, η πιθανότητα να υπάρχει σύνδεση μεταξύ τους θα είναι ίση με $p = \frac{150}{276} \approx 0.5434$.

Γνωρίζοντας τώρα την πιθανότητα σύνδεσης δύο κόμβων στον γράφο μπορούμε να πούμε ότι αναμένουμε κατά μέσο όρο ο συντελεστής ομαδοποίησης ενός κόμβου να είναι περίπου ίσος με $0.5434$, που όπως βλέπουμε, στην περίπτωσή μας βρίσκουμε ότι ο μέσος όρος του συντελεστή ομαδοποίησης είναι ίσος με $0.5642$.

- <b>Τοπολογία Random Geometric Graph: </b> Σε αυτή την περίπτωση τοπολογίας, αρχικά επιλέγονται $n$ το πλήθος σημεία στον κύβο διάστασης $d$. Η σύνδεση δύο κόμβων καθορίζεται αν η απόστασή τους είναι μικρότερη ή ίση μιας ακτίνας $R$. Στην περίπτωσή μας, έχουμε $n=24, d=2$ και $R = 0.35$. Όπως βλέπουμε, στην περίπτωσή μας ο μέσος συντελεστής ομαδοποίησης είναι ίσος με $0.5933$ και η διασπορά ίση με $0.0637$.

- <b>Τοπολογία Scale Free: </b> Η τοπολογία αυτή κατασκευάζεται μέσω του αλγορίθμου $\text{Barab\'asi - Albert}$. Ο αλγόριθμος δέχεται δύο παραμέτρους: $n$ (24 στην συγκεκριμένη περίπτωση) που αντιστοιχεί στο πλήθος των κόμβων του τελικού γράφου και $m$ (4 στη δική μας περίπτωση) τον αριθμό των συνδέσεων που γίνονται με την εισαγωγή ενός νέου κόμβου. Αρχικά αρχικοποιείται ένας γράφος με $m+1$ κόμβους. Ένας κόμβος επιλέγεται τυχαία και συνδέεται τυχαία με όλους τους υπολοίπους $m$. Εν συνεχεία, σε κάθε επανάληψη του αλγορίθμου προστίθεται ένας νέος κόμβος στον γράφο ο οποίος συνδέεται με τους $m$ κόμβους $i$ που έχουν τη μεγαλύτερη τιμή στην ποσότητα 

$$\Pi(k_i) = \frac{k_i(t)}{\sum_{j}k_j(t)},$$

όπου $k_i(t)$ είναι ο βαθμός του κόμβου $i$ κατά την $t$-οστή επανάληψη και το άθροισμα εκτείνεται προς όλους τους κόμβους του γράφου τη δεδομένη χρονική στιγμή. Η κατανομή του βαθμού των κόμβων αυτών των γράφων είναι μια Power-law κατανομή όπου η πιθανότητα ένας κόμβος να έχει βαθμό $k$ είναι ανάλογη της $\mathbb{P}(k) = k^{-\gamma}$, με $\gamma \in [2.1, 4]$ βάση εμπειρικών αποτελεσμάτων. Χαρακτηριστικό αυτών των γράφων είναι ότι υπάρχουν λίγοι κόμβοι με υψηλό βαθμό κόμβου (κόμβοι hub) ενώ οι περισσότεροι κόμβοι έχουν χαμηλό βαθμό. Έτσι, αντίστοιχη συμπεριφορά αναμένουμε και απ' την κατανομή του συντελεστή ομαδοποίησης, πράγμα το οποίο βλέπουμε και απ' το 4ο διάγραμμα κατά σειρά όπου οι περισσότεροι κόμβοι έχουν συντελεστή ομαδοποίησης ιδιαίτερα χαμηλό που κυμαίνεται μεταξύ $0.2-0.5$. Αυτό φαίνεται και απ' τη μέση τιμή του συντελεστή ομαδοποίησης που είναι ιδιαίτερα χαμηλός και ίσος με $0.3134$.

- <b> Τοπολογία Small World: </b> Στη συγκεκριμένη περίπτωση ο αλγόριθμος των Watts - Strogratz παράγει έναν γράφο με τοπολογία Small World. Ο αλγόριθμος λειτουργεί ως εξής: Αρχικά τοποθετούνται οι $n (=24)$ κόμβοι στο επίπεδο σε κυκλική διάταξη. Κάθε κόμβος $i$ έχει $k (=4)$ γείτονες τους οποίους συμβολίζουμε με το σύνολο $\{i-\frac{k}{2},\dots,i-1,i+1,\dots, i+\frac{k}{2}\}$, όπου ο συμβολισμός $i-h$ υποδηλώνει τον κόμβο που βρίσκεται $h$ θέσεις αντίστροφα της φοράς του ρολογιού στον δακτύλιο ενώ το $i+h$ υποδηλώνει τον κόμβο που βρίσκεται $h$ θέσεις ως προς την φορά του ρολογιού απ' τον κόμβο $i$. Ο αλγόριθμος επαναλαμβάνεται για $\frac{k}{2}$ γύρους, καθένας απ' τους οποίους αποτελείται από $n$ περαιτέρω υπο-γύρους (επαναλήψεις). Σε έναν γύρο $j$, για κάθε έναν απ' τους υπογύρους, ένας κόμβος $i$ του γράφου επιλεγέται μαζί με την ακμή που συνδέεται με τον κόμβο-γείτονα $i+j$ (με τη φορά του ρολογιού). Τότε, με πιθανότητα $p (=0.25)$ η ακμή $(i,i+j)$ αποσυνδέεται από τον κόμβο $i+j$ και συνδέεται με έναν τυχαία ομοιόμορφα επιλεγμένο κόμβο του πλέγματος. 

Έτσι, στο τέλος καταλήγουμε με έναν γράφο Small World τοπολογίας. Οι γράφοι αυτοί χαρακτηρίζονται από σχετικά μικρό μέσο μήκους μονοπατιού. Όσον αφορά την κατανομή του βαθμού τον κόμβων όπως και του συντελεστή ομαδοποίησης αυτή δεν έχει κάποια συγκεκριμένη μορφή απαραίτητα. Στο παράδειγμά μας έχουμε $n=24, k=4$ και $p=0.25$. Βλέπουμε ότι ο μέσος συντελεστής ομαδοποίησης είναι ιδιαίτερα χαμηλός και ίσος με $0.3127$ με αντίστοιχη διασπορά $0.0520$. Βλέποντας την κατανομή απ' το τελευταίο σχήμα της παραπάνω εικόνας μπορούμε να δούμε ότι οι περισσότεροι κόμβοι έχουν χαμηλή τιμή συντελεστή ομαδοποίησης που κυμαίνεται μεταξύ $0.1 - 0.5$ ενώ υπάρχει ένας κόμβος (ο 15ος) με τιμή 1.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    connected_comp = list(nx.connected_components(G))
    distances = []
    diameter,radius = 0,0
    for component in connected_comp: # Split in each connected component
        G_sub = G.subgraph(component)
        diameter = max(diameter,nx.diameter(G_sub))
        radius = max(radius, nx.radius(G_sub))
        d = dict(nx.all_pairs_shortest_path_length(G_sub))
        for node1, d2 in d.items():
            for node2, val in d2.items():
                if node1 != node2:
                    distances.append(val)
    if not nx.is_connected(G):
        radius = np.inf
    return np.mean(distances), np.var(distances), diameter, radius # Returns mean_path, var_path, diameter, radius
```

```python
for G in graphs:
    res = compute_path_lengths(G)
    print(f"- " + G.graph['name'] + f": Mean length path: {res[0]}, Var length path: {res[1]}, Diameter: {res[2]}, Radius: {res[3]}")
```

Αυτό το αποτέλεσμα συμφωνεί με τη θεωρία για αυτές τις τοπολογίες καθώς χαρακτηριζόνται από μεγάλο μέσο μήκος μονοπατιού. Επίσης, όπως είναι λογικό, ο γράφος αυτός έχει και τη μεγαλύτερη διάμετρο η οποία ισούται με την ακτίνα του. Απ' την άλλη, το μικρότερο μέση μήκος μονοπατιού το έχει ο γράφος που αντιστοιχεί στην τοπολογία τύπου Scale-Free και ο τυχαίος γράφος $\text{Erd\H{o}s - R\'enyi}$. Στην περίπτωση του τυχαίου γεωμετρικού γράφου έχουμε ακτίνα ίση με την τιμή άπειρο αφού ο γράφος που προκύπτει είναι μη συνεκτικός. Για αυτόν τον γράφο, ορίζουμε ως διάμετρο τη μεγαλύτερη διάμετρο από όλες τις συνεκτικές συνιστώσες. Έτσι, συμπεραίνουμε ότι η διάμετρος της μεγαλύτερης συνεκτικής συνιστώσας απ' τις δύο είναι και η διάμετρος του γράφου που έχει τιμή $5$.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
def find_nodes(G):
    diameter = 0 
    subgraphs = []
    results = {}
    for component in list(nx.connected_components(G)): # Calculating the diameter, need to consider all possible connected components
        G_sub = G.subgraph(component)
        diameter = max(diameter, nx.diameter(G_sub))
        subgraphs.append(G_sub)
    for G_sub in subgraphs:
        distances = dict(nx.all_pairs_shortest_path_length(G_sub))
        for node1, d in distances.items():
            results[node1] = []
            for node2, val in d.items():
                if val == diameter:
                    results[node1].append(node2)
    max_nodes = max([len(l) for l in results.values()])
    final_results = {}
    for node,l in results.items():
        if len(l) == max_nodes:
            final_results[node] = l
    return final_results, diameter, max_nodes
```

```python
for G in graphs:
    print(f"{5*'-'}> Results for " + G.graph['name'] + f" graph <{5*'-'}\n")
    res, diameter, max_nodes = find_nodes(G)
    print(f"- Diameter: {diameter}")
    print(f"- Max number of nodes that the shortest path is equal to the diameter: {max_nodes}")
    for k,v in res.items():
        print(f"- Node {k} has shortest path length equal to the diameter ({diameter}) when the end node(s) are: {v}")
    print("\n")
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    periphery = []
    diameter = 0
    for comp in list(nx.connected_components(G)):
        G_sub = G.subgraph(comp)
        diameter = max(diameter, nx.diameter(G_sub))
    for comp in list(nx.connected_components(G)):
        G_sub = G.subgraph(comp)
        eccentricities = nx.eccentricity(G_sub)
        for node,eccen in eccentricities.items():
            if eccen == diameter: periphery.append(node)
    return periphery
```

```python
def find_center(G):
    center = []
    radius = 0
    for comp in list(nx.connected_components(G)):
        G_sub = G.subgraph(comp)
        radius = max(radius, nx.radius(G_sub))
    for comp in list(nx.connected_components(G)):
        G_sub = G.subgraph(comp)
        eccentricities = nx.eccentricity(G_sub)
        for node,eccen in eccentricities.items():
            if eccen == radius: center.append(node)
    return center
```

```python
print(f"{10*'-'}> Results <{10*'-'}\n")
for G in graphs:
    print(f"{5*'-'} > Periphery and center of " + G.graph["name"] + f" graph< {5*'-'}\n")
    periphery = find_periphery(G)
    center = find_center(G)
    print(f"- Periphery: {periphery}")
    print(f"- Center: {center}\n")
```

Υπολογίζοντας βλέπουμε ότι η πιθανότητα ένας κόμβος να έχει βαθμό 4 είναι ίση με $\mathbb{P}(4) = \frac{10}{26}\approx 0.3846$. Έτσι, σε αυτή την περίπτωση, με βάση το σχήμα μπορούμε να πούμε ότι η κατανομή των κόμβων είναι η Poisson με μέση τιμή $\lambda \approx 0.3846.$

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G, w = 4, h = 2):
    a = 1/(max(nx.adjacency_spectrum(G)).real)-0.01
    katz = list(nx.katz_centrality(G, alpha = a).values())
    print(f"{5*'-'} > Results for " + G.graph['name'] + f" graph <{5*'-'}")
    fig,ax = plt.subplots(figsize = (w,h))
    ax.hist(katz)
    ax.set_title(f"Distribution of Katz centralities in " + G.graph['name'] + " graph")
    ax.set_ylabel("Frequency")
    ax.set_xlabel("Katz centrality")
    plt.show()
    print(f"\n- Mean node Katz centrality: {np.mean(katz)}")
    print(f"- Variance node Katz centrality: {np.var(katz)}\n")
```

```python
for G in graphs:
    find_katz(G)
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    results = {}
    # Degree
    degrees = dict(nx.degree(G))
    degrees = list(zip(degrees.values(), degrees.keys()))
    max_degree = max(degrees)[0]
    results["Degree"] = [node for v,node in degrees if v == max_degree]
    
    # Closeness centrality
    closeness = dict(nx.closeness_centrality(G))
    closeness = list(zip(closeness.values(), closeness.keys()))
    max_closeness = max(closeness)[0]
    results["Closeness centrality"] = [node for v,node in closeness if v == max_closeness]
    
    # Betweeness centrality
    betweeness = dict(nx.betweenness_centrality(G))
    betweeness = list(zip(betweeness.values(), betweeness.keys()))
    max_betweeness = max(betweeness)[0]
    results["Betweeness centrality"] = [node for v,node in betweeness if v == max_betweeness]
    
    # Katz centrality
    a = 1/(max(nx.adjacency_spectrum(G)).real)-0.01
    katz = dict(nx.katz_centrality(G, alpha = a))
    katz = list(zip(katz.values(), katz.keys()))
    max_katz = max(katz)[0]
    results["Katz centrality"] = [node for v, node in katz if v == max_katz]
    
    return results
```

```python
for G in graphs:
    results = find_central(G)
    print(f"{5*'-'}> Results for " + G.graph['name'] + f" graph <{5*'-'}")
    print("- Nodes with maximum degree: ", results["Degree"])
    print("- Nodes with maximum closeness centrality: ", results["Closeness centrality"])
    print("- Nodes with maximum betweeness centrality: ", results["Betweeness centrality"])
    print("- Node with maximum Katz centrality: ", results["Katz centrality"])
    print("\n")
```

#### Δ1.3 Ποιες είναι οι κρίσιμες τιμές για κάθε τύπο δικτύου οι οποίες οδηγούν σε μετάβαση φάσης (phase transitions);
#### Δ1.4 Ποιες μεταβάσεις φάσης είναι απότομες (sharp) και ποιες ομαλές (smooth);

```python
# A custom function to generate the diagrams

def generate_graph(type, n, p, prob = None):
    if type == "REG":
        return nx.watts_strogatz_graph(n=n,k=p,p=0.0)
    elif type == "RGER":
        return nx.gnm_random_graph(n,p)
    elif type == "RG-Gilbert":
        return nx.gnp_random_graph(n,p)
    elif type == "RGG":
        return nx.random_geometric_graph(n,p)
    elif type == "SF":
        return nx.barabasi_albert_graph(n,p)
    elif type == "SW":
        return nx.watts_strogatz_graph(n = n, k = p, p = prob)

def generate_diagram(type):
    configs = {"REG": {"step": 2, "l_bound": 2, "u_bound": 10, "parameter": "d"},
               "RGER": {"step": 100, "l_bound": 100, "u_bound": 800, "parameter": "M"},
               "RG-Gilbert": {"step": 0.01, "l_bound": 0.01, "u_bound": 0.1, "parameter": "p"},
               "RGG": {"step": 0.025, "l_bound": 0.025, "u_bound": 0.25, "parameter": "R"},
               "SF": {"step": 2, "l_bound": 2, "u_bound": 10, "parameter": "d"},
               "SW": {"step1": 2, "l1_bound": 2, "u1_bound": 10, "step2": 0.1, "l2_bound": 0.1,
                      "u2_bound": 0.7, "parameter1": "d", "parameter2": "prob"}}
    
    if type == "SW":
        step1 = configs[type]["step1"]
        l1_bound = configs[type]["l1_bound"]
        u1_bound = configs[type]["u1_bound"]
        step2 = configs[type]["step2"]
        l2_bound = configs[type]["l2_bound"]
        u2_bound = configs[type]["u2_bound"]
        param1 = configs[type]["parameter1"]
        param2 = configs[type]["parameter2"]
        
        param1_range = np.arange(l1_bound, u1_bound+step1, step1) # Num neighbors
        param2_range = np.arange(l2_bound, u2_bound+step2, step2) # Probability
        
        fig, ax = plt.subplots(figsize = (14, 10), ncols = len(param1_range),
                               sharey = True, nrows = 2)
        for i, k in enumerate(param1_range):
            for m, n in enumerate((100, 200)):
                results = []
                for p in param2_range:
                    succ = 0 
                    for j in range(100):
                        if nx.is_connected(generate_graph(type, n, k, p)):
                            succ +=1 
                    results.append(succ/100)
                ax[m, i].plot(param2_range, results)
                ax[m, i].set_title("d = " +f"{k}, n = {n}", fontsize = 10)
                ax[m, i].set_xlabel(param2)
                ax[m, i].set_ylabel("Connectivity ratio")
        fig.suptitle("Connectivity " + type)
        plt.show()   
    else:
        step = configs[type]["step"]
        l_bound = configs[type]["l_bound"]
        u_bound = configs[type]["u_bound"]
        param = configs[type]["parameter"]
        
        for n in (100,200):
            results = []
            parameter_range = np.arange(start = l_bound, stop = u_bound+step, step = step)
            for p in parameter_range:
                successess = 0
                for i in range(100):
                    if nx.is_connected(generate_graph(type,n,p)):
                        successess+=1
                results.append(successess/100)
            plt.plot(parameter_range, results, label = "n="+str(n))
        plt.title("Connectivity " + type)
        plt.legend()
        plt.xlabel(param)
        plt.ylabel("Connectivity Ratio")
        plt.grid()
        plt.show()
            
```

```python
print("Connectivity REG")

generate_diagram(type = "REG")
```

#### Να σχολιαστούν τα αποτελέσματα.

```python
def find_cut(G):
    if nx.is_connected(G):
        return str(len(nx.minimum_node_cut(G)))
    else:
        return f"The graph is already disconnected."
```
